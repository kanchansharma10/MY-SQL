    CRUD Operations in  My-sql <br>
    <h3 style="text-align: center; color: white; font-size: 15px; font-family: 'Arial', sans-serif;"></h3>
</div>

CRUD operations in **MySQL** represent the **basic operations** used to interact with a database. CRUD stands for:

* **C**reate – Insert data into a table
* **R**ead – Retrieve data from a table
* **U**pdate – Modify existing data
* **D**elete – Remove data from a table

Here's a breakdown of each operation with SQL syntax and an example:

---

###  1. **CREATE** (Insert Data)

**Syntax:**

```sql
INSERT INTO table_name (column1, column2, ...)
VALUES (value1, value2, ...);
```

**Example:**

```sql
INSERT INTO users (name, email)
VALUES ('Alice', 'alice@example.com');
```

---

###  2. **READ** (Retrieve Data)

**Syntax:**

```sql
SELECT column1, column2, ...
FROM table_name
WHERE condition;
```

**Example:**

```sql
SELECT * FROM users;
```

Or with a condition:

```sql
SELECT name, email FROM users
WHERE id = 1;
```

---

###  3. **UPDATE** (Modify Data)

**Syntax:**

```sql
UPDATE table_name
SET column1 = value1, column2 = value2, ...
WHERE condition;
```

**Example:**

```sql
UPDATE users
SET email = 'newalice@example.com'
WHERE id = 1;
```

---

###  4. **DELETE** (Remove Data)

**Syntax:**

```sql
DELETE FROM table_name
WHERE condition;
```

**Example:**

```sql
DELETE FROM users
WHERE id = 1;
```

**Important:** Always use `WHERE` with `UPDATE` and `DELETE` to avoid modifying/deleting all records.
